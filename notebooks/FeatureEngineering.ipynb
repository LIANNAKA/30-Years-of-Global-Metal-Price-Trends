```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/processed/cleaned_metal_prices.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df.sort_values("Date")
df.head()
```

```python
df["Year"] = df["Date"].dt.year
df["Month_Num"] = df["Date"].dt.month
df["Quarter"] = df["Date"].dt.quarter
df.head()
```

For Gold, Silver, Aluminium, Nickel, Uranium

```python
metals = ["Price_gold", "Price_silver", "Price_alum", "Price_nickel", "Price_uran" ]
```

Lag Features = previous values,

Rolling mean healps to smooth out any fluctuations or sudden spkies,

Volatality , percentage change

```python
for metal in metals:

    #Lag Features
    df[f"{metal}_Lag_1"] = df[metal].shift(1)
    df[f"{metal}_Lag_2"] = df[metal].shift(2)
    df[f"{metal}_Lag_3"] = df[metal].shift(3)

    #Rolling Mean
    df[f"{metal}_Rolling_Mean_3"] = (df[metal].rolling(window=3).mean())

    #Percentage Returns
    df[f"{metal}_Returns"] = (df[metal].pct_change())
    

```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.to_csv("../data/processed/feature_engineering_data.csv", index=False)
```
