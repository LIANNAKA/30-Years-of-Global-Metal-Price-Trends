```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('MetalPriceChange.csv')
```

```python
data = data.dropna(how="all")
```

```python
data['Price_uran'] = data['Price_uran'].interpolate()
data['Price_uran'] = data['Price_uran'].ffill()
```

```python
data.isnull().sum()
```

```python
data['Inflation_rate'] = data['Inflation_rate'].interpolate()
data['Inflation_rate'] = data['Inflation_rate'].ffill()
data['Price_alum_infl'] = data['Price_alum_infl'].interpolate()
data['Price_alum_infl'] = data['Price_alum_infl'].ffill()
data['Price_gold_infl'] = data['Price_gold_infl'].interpolate()
data['Price_gold_infl'] = data['Price_gold_infl'].ffill()
data['Price_uran_infl'] = data['Price_uran_infl'].interpolate()
data['Price_uran_infl'] = data['Price_uran_infl'].ffill()
data['Price_silver_infl'] = data['Price_silver_infl'].interpolate()
data['Price_silver_infl'] = data['Price_silver_infl'].ffill()
data['Price_nickel_infl'] = data['Price_nickel_infl'].interpolate()
data['Price_nickel_infl'] = data['Price_nickel_infl'].ffill()
```

```python
data = data.dropna(subset=["Year"])
```

```python
data["Year"] = data["Year"].astype(int)
```

```python
data["Date"] = pd.to_datetime(
    data["Year"].astype(str) + "-" + data["Month"],
    errors="coerce"
)
```

```python
data.set_index("Date", inplace=True)
```

```python
plt.figure(figsize=(10,5))
plt.plot(data.index, data["Price_gold"])
plt.title("Gold Price Trend Over Time")
plt.xlabel("Year")
plt.ylabel("Gold Price")
plt.show()
```

```python
data['Price_silver'].plot(kind="box", title="Metal Price Volatility")
```

```python
price_cols = ["Price_alum", "Price_gold","Price_silver"]
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14,10))

# Aluminium
data[["Price_alum", "Price_alum_infl"]].plot(ax=axes[0,0], title="Aluminium")

# Gold
data[["Price_gold", "Price_gold_infl"]].plot(ax=axes[0,1], title="Gold")

# Silver
data[["Price_silver", "Price_silver_infl"]].plot(ax=axes[1,0], title="Silver")

# Uranium
data[["Price_uran", "Price_uran_infl"]].plot(ax=axes[1,1], title="Uranium")

plt.tight_layout()
plt.show()
```
